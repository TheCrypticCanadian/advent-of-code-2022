```python
import math
class Monkey:
    def __init__(self,name,starting_items,operation,test,test_true,test_false):
        #print(test,test_true,test_false)
        self.name = name
        self.items = list(map(int,starting_items.split(":")[1].split(",")))
        self.operation = operation.split(":")[1].split("=")[1].strip()
        self.test = int(test.split(":")[1].strip().split(" ")[2])
        self.test_true = int(test_true.split(":")[1].strip().split(" ")[3])
        self.test_false = int(test_false.split(":")[1].strip().split(" ")[3])
        self.total_inspections = 0
        self.print = False

    def print(self):
        print()
        print("Monkey Name: ",self.name)
        print("Items: ",self.items)
        print("Operation: ",self.operation)
        print("Test: ",self.test)
        print("Test True: ", self.test_true)        
        print("Test False: ",self.test_false)

    def inspect_item(self):
        item = self.items.pop(0)
        if self.print:
            print(f"\tMonkey inspects an item with a worry level of {item}.")
        self.total_inspections += 1
        return item
    
    def calc_worry(self,item):
        old = int(item)
        worry = eval(self.operation)

        op_strs = self.operation.split(" ")
        if op_strs[1] == "*":
            action = "multiplies"
        else:
            action = "adds"

        if self.print:
            print(f"\t\tWorry level is {action} by {op_strs[2]} to {worry}.")

        return worry
    
    def bored(self,item):
        old = int(item)
        worry = int(math.floor(item/3))
        if self.print:
            print(f"\t\tMonkey gets bored with item. Worry level is divided by 3 to {worry}.")
        return worry

    def test_item(self,item):
        if item % self.test == 0:
            if self.print:
                print(f"\t\tCurrent worry level is divisible by {self.test}.")
            return self.test_true
        else:
            if self.print:
                print(f"\t\tCurrent worry level is not divisible by {self.test}.")
            return self.test_false

    def catch_item(self,item):
        if self.print:
            print(f"\t\tItem with worry level {item} is thrown to monkey {self.name}.")
        self.items.append(item)



monkeys = {}
inspect_totals = {}
for monkey_in in monkeys_in:
    monkeys[int(monkey_in[0][7])] = Monkey(int(monkey_in[0][7]),monkey_in[1],monkey_in[2],monkey_in[3],monkey_in[4],monkey_in[5])
    inspect_totals[int(monkey_in[0][7])] = 0


for round in range(20):
    print()
    print(f"Round {round+1}: ")
    for monkey_index in monkeys:

        #print(f"Monkey {monkey_index}:")
        monkey = monkeys[monkey_index]
        while len(monkey.items) != 0:
            item = monkey.inspect_item()
            worry_item = monkey.calc_worry(item)
            bored_item = monkey.bored(worry_item)
            test_results = monkey.test_item(bored_item)
            monkeys[test_results].catch_item(bored_item)

    for monkey_index in monkeys:
        inspect_totals[monkey_index] += len(monkeys[monkey_index].items)
        print(f"Monkey {monkey_index}: ",monkeys[monkey_index].items)

inspection_values = []
for monkey_index in monkeys:
    total_i = monkeys[monkey_index].total_inspections
    print(f"Monkey {monkey_index} inspected items {total_i} times.")
    inspection_values.append(total_i)

inspection_values = sorted(inspection_values)
inspection_values.reverse()
print(inspection_values)
print(inspection_values[0]*inspection_values[1])




        
    
```
