```python
def dijkstra(map,source_char,end_char):
    def get_neighbours(position):
        up = (position[0]-1,position[1])
        down = (position[0]+1,position[1])
        left = (position[0],position[1]-1)
        right = (position[0],position[1]+1)
        return [up,down,left,right]

    # Get Source Position
    source_position = ()
    for y in range(len(maps)):
        try:
            start_index = maps[y].index(source_char)
        except:
            start_index = -1
        if start_index != -1:
            source_position = (y,start_index)
    #print(source_position)
    # Get End Position
    end_position = ()
    for y in range(len(maps)):
        try:
            end_index = maps[y].index(end_char)
        except:
            end_index = -1
        if end_index != -1:
            end_position = (y,end_index)
    #print(end_position)

    # Initial distance dict
    distance_dict = {}
    prev_dict = {}
    char_map = {}
    elevation = {}
    Q = []
    min_y = 0 
    min_x = 0
    max_y = len(map)
    max_x = len(map[0])
    for y in range(len(map)):
        for x in range(len(map[y])):
            distance_dict[(y,x)] = float('inf')
            prev_dict[(y,x)] = None
            char_map[(y,x)] = map[y][x]
            Q.append((y,x))

    # Set source to 0
    distance_dict[source_position] = 0
    elevation[source_position] = 0

    current_distance = 0
    while len(Q) > 0:

        distance_vals = [distance_dict[tuple] for tuple in Q]
        u = Q.pop(distance_vals.index(min(distance_vals)))
        neighbours = get_neighbours(u)

        
        #print(neighbours)
        for v in neighbours:
            if (v in Q):
                if char_map[v] == "E":
                    elevation[v] = ord("z")-96
                elif char_map[v] == "S":
                    elevation[v] = ord("a")-96
                else:
                    elevation[v] = ord(char_map[v])-96
                if (elevation[v] <= elevation[u]+1):
                    alt = distance_dict[u]+1
                    if alt < distance_dict[v]:
                        distance_dict[v] = alt
                        prev_dict[v] = u
    print(distance_dict[end_position])

    shortest = []
    u = end_position
    if prev_dict[u] != None:
        while prev_dict[u]:
            shortest.append(prev_dict[u])
            u = prev_dict[u]
    #print(shortest)


path = dijkstra(maps,"S","E")
```
